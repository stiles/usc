### BONUS: Group by name and year for just surnames in our class, aggregating with the mean per_100k value

```python
students_name = df[df.name.isin(students)]
```

```python
students_name.groupby(["name"]).agg({"per_100k": "mean"}).round(
    2
).reset_index().sort_values("per_100k", ascending=False)
```
