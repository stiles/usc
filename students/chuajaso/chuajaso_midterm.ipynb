```python
# Reported Crimes
import pandas as pd

pd.set_option("display.max_columns", None)

%load_ext lab_black
```

```python
df = pd.read_csv("./data/Crime_Data_from_2020_to_Present.csv")
df.columns
```

```python
# What crimes involve rental motor scooters?
df[df["Mocodes"].str.contains("0947", na=False)]
```

```python
# I want a single crime code col
df["CRIME_CODES"] = (
    df["Crm Cd 1"].astype(str)
    + " "
    + df["Crm Cd 2"].astype(str)
    + " "
    + df["Crm Cd 3"].astype(str)
    + " "
    + df["Crm Cd 4"].astype(str)
)
df.head()
```

```python
# Interesting. I would expect more.
# How does that compare to bikes?
# 480 (Bicycle - stolen)
# 485 (Bicycle - attempted stolen)
df[
    df["CRIME_CODES"].str.contains("480", na=False)
    | df["CRIME_CODES"].str.contains("485", na=False)
]
```

```python
# Focus on central
central = df[df["AREA NAME"] == "Central"]
central
```

```python
# What about bicycles in central?
central[
    central["CRIME_CODES"].str.contains("480", na=False)
    | central["CRIME_CODES"].str.contains("485", na=False)
]
```

```python
# 1 in 10 bike theft attempts occur in central LA
# Where do these bike thefts occur?
central[
    central["CRIME_CODES"].str.contains("480", na=False)
    | central["CRIME_CODES"].str.contains("485", na=False)
].value_counts("Premis Cd").head()
```

```python
# Seems to be a large jump at 102. What are those?
central_bikes = central[
    central["CRIME_CODES"].str.contains("480", na=False)
    | central["CRIME_CODES"].str.contains("485", na=False)
]
central_bikes[central_bikes["Premis Cd"] == 102]
```

```python
# Sidewalks! Makes sense. How many bike thefts in total in central?
len(central_bikes)
```

```python
# This is the percentage of bikes in central la stolen in sidewalks?
85 / len(central_bikes)
```

```python
# Nearly 1 in 5 bike thefts in central la occur in sidewalks!
```
