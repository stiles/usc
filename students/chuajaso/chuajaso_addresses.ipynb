```python
import pandas as pd

%load_ext lab_black
```

```python
# Import data
df = pd.read_csv("./data/Addresses_in_the_City_of_Los_Angeles.csv")
df.head()
```

```python
# This is a list of all addresses in LA City. Thanks LA Open Data Portal (https://data.lacity.org/City-Infrastructure-Service-Requests/Addresses-in-the-City-of-Los-Angeles/4ca8-mxuh)
# Remove non-interesting or non-useful data.
df = df.drop(columns=["PIND", "LAT", "LON", "X_COORD_NBR", "Y_COORD_NBR"])
df.head()
```

```python
# How many zip codes serve LA?
len(df["ZIP_CD"].unique())
```

```python
# Which zip code has the most addresses?
df.value_counts("ZIP_CD")
```

```python
# Interesting. There are zip codes with only one address. Is this common?
zip_codes = df.value_counts("ZIP_CD")
zip_codes[zip_codes == 1]
```

```python
# Okay. Looks like there are only 5. Where are these places?
df[df["ZIP_CD"].isin(zip_codes[zip_codes == 1].index.tolist())]
```

```python
# I can't see anything interesting with the zip codes. How many addresses are odd?
len(df[df["HSE_NBR"] % 2 == 1])
```

```python
# Even?
len(df[df["HSE_NBR"] % 2 == 0])
```

```python
# How many addresses share my zip code?
len(df[df["ZIP_CD"] == 90007])
```
