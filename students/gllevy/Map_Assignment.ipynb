```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
pd.options.display.max_columns = 1000
pd.options.display.max_rows = 1000
```

```python
elections_df = pd.read_csv(
    "https://raw.githubusercontent.com/stiles/notebooks/master/elections/data/ca_counties_election_results_2020.csv",
    dtype={"county_fips": str},
)
```

```python
counties = gpd.read_file(
    "https://raw.githubusercontent.com/stiles/usc/main/data/geo/counties_simple.geojson",
    dtype={"county_fips": str},
)
```

```python
race_la_df = pd.read_csv(
    "https://raw.githubusercontent.com/stiles/usc/main/data/processed/la_neighborhood_demographics.csv"
)
```

```python
elections_df.head()
```

```python
race_la_df.head()
```

```python
counties.head()
```

```python
elections = elections_df.drop("county_name", axis=1)
```

```python
len(elections_df)
```

```python
len(race_la_df)
```

```python
# Join!
```

```python
counties_elections = pd.merge(counties, elections, on="county_fips")
```

```python
counties_elections[counties_elections["county_name"] == "Claremont"]
```

```python
race_la_df = race_la_df.rename(columns={"name": "county_name"})
```

```python
race_la_df[race_la_df["county_name"] == "Claremont"]
```

```python
race_elections = pd.merge(race_la_df, counties_elections, on="county_name")
```

```python
race_elections.head()
```

```python
counties_elections.to_file("./data/election_map.json", driver="GeoJSON")
```
