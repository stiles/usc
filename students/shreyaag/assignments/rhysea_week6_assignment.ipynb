### Part II: Exploring Dataset 'conc'

```python
#I'm making the assumption that shorter buildings are less in danger than taller buildings are. 
#So, I am organizing the data according to number of stories to indicate danger levels.
less_danger = conc.sort_values("stories",ascending=True)
```

```python
less_danger.head()
```

```python
#Conversely, taller buildings are not only in greater danger, but more dangerous to the surrounding areas too. 
#Hence, I am sorting the data based on most dangerous buildings.
more_danger = conc.sort_values("stories",ascending=False)
```

```python
more_danger.head()
```

```python
#I want to find out the number of buildings that have a particular number of stories
#I am running value_counts on the stories column to create a dataset that categories on the basis of no. of stories.
stories_df = conc['stories'].value_counts()
stories_df
```

```python
#I want to find out the average height/number of stories in the dataset
#Gives an interesting insight into risk and hazards from earthquakes
stories_des = more_danger["stories"].describe()
stories_des
```

### Part VI: Merging Datasets

```python
economics = economics.rename(columns={"id":"zip_code"})
```

```python
economics.head()
```

```python
merge_df = pd.merge(zips_count, economics, on=["zip_code"])
```

```python
merge_df.head()
```

```python
merge_df=merge_df.rename(columns={"num":"buildings_at_risk"})
```

```python
merge_df.sort_values("buildings_at_risk",ascending=False)
```

```python
final_df = merge_df[["zip_code","buildings_at_risk","name","totpop_cy","wlthrnk_cy","tothh_cy"]]
final_df.head()
```

```python
final_df=final_df.rename(columns={"totpop_cy":"total_pop","wlthrnk_cy":"wealth_rank","tothh_cy":"total_households"})
```

```python
final_df.sort_values("buildings_at_risk",ascending=False)
```
