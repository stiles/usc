# Get data from HTML tables

```python
import pandas as pd
import altair as alt
import numpy as np
```

### Get url

```python
url = "https://www.ssa.gov/oact/babynames/numberUSbirths.html"
```

### Read it into a dataframe

```python
births = pd.read_html(url)[0]
```

### Clean up columns

```python
births.rename(
    columns={
        "Year of birth": "year",
        "Male": "male",
        "Female": "female",
        "Total": "total",
    },
    inplace=True,
)
```

### Convert year

```python
births.year = births.year.astype(str)
```

### Chart it

```python
alt.Chart(births).mark_bar(color='red').encode(
    x=alt.X('year:T', axis=alt.Axis(tickCount=20)),
    y='total'
).properties(title='Births by year', width=900)
```

```python
births.to_csv('../data/processed/ssa_births_1880_2020.csv', index=False)
```
