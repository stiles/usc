# 1.Use groupby for investigating age information

```python
df.dropna(subset = ["Age"]) # Filtering the Age data to delete Null
```

```python
df_Agefiltering = df.dropna(subset = ["Age"])
```

```python
df_Agefiltering [df_Agefiltering["Survived"] == 1 ] # Filtering the number of surviors with valid data age
```

```python
df_Agefiltering_survived = df_Agefiltering [df_Agefiltering["Survived"] == 1 ]
```

The number of survivors from different class cabins

```python
df_Agefiltering_survived.groupby("Pclass").size()
```

The average age of survivors or non-survivors in different class cabins 

```python
df_Agefiltering.groupby(["Pclass", "Survived"])["Age"].mean()
```

Median, mode, and mean ages of survivors in different class cabins  

```python
df_Agefiltering_survived.groupby(["Pclass"])["Age"].agg(["min","max","mean"]) 

# The reason of min below 1 is because of infants
```

# 2.Use groupby for investigating sex information

The number of survivors or non-survivors classified by gender

```python
df.groupby("Sex")["Survived"].value_counts()
```

The average fare classified by Sex in different class cabins

```python
df.groupby(["Pclass","Sex"]).agg({"Fare":"mean"}).reset_index()
```

The number of survivors or non-survivors classified by gender in different class cabins

```python
df.groupby(["Pclass", "Survived"])["Sex"].value_counts()
```

# 3.Use groupby for investigating fare information

The average fare for each class cabins

```python
df.groupby(["Pclass"]).agg({"Fare": "mean"}).round(2).reset_index()
```

The average fare from each port of embarkation

```python
df.dropna(subset = ["Embarked"]) # Filtering the Embarked data to delete Null
```

```python
df_embarkedfiltering = df.dropna(subset = ["Embarked"])
```

```python
df_embarkedfiltering.groupby(["Embarked"]).agg({"Fare": "mean"}).round(2).reset_index()
```
