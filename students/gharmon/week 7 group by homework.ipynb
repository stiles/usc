### Group colors by the number of times they have each been reported

```python
ufo_df.groupby("Colors Reported").count()
```

### Group shapes by the number of times they have each been reported

```python
ufo_df.groupby("Shape Reported").count()
```

### Group by state and the "Fireball" shape and find what states reported the most "Fireball" UFOS

```python
fireball_ufos = ufo_df[ufo_df["Shape Reported"] == "FIREBALL"]
```

```python
fireball_ufos.groupby(["State"]).count().sort_values(
    "Shape Reported", ascending=False
).head()
```

### In Seattle, Washington, what were the most common colors and shapes

```python
seattle_ufos = ufo_df[ufo_df["City"] == "Seattle"]
```

```python
seattle_ufos.groupby(["Shape Reported"]).count().sort_values(
    "Shape Reported", ascending=False
).head()
```

```python
seattle_ufos.groupby(["Colors Reported"]).count().sort_values(
    "Colors Reported", ascending=False
).head()
```
