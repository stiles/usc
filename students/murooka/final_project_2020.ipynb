#### Load python tools

```python
import pandas as pd
import geopandas as gpd
```

```python
pd.options.display.max_columns = 1000
pd.options.display.max_rows = 1000
```

---

#### Read CSV value file

```python
covid_19_df = pd.read_csv("data/covid19cases_test.csv")
```

#### Read First 5 Rows

```python
covid_19_df.head()
```

#### Rename Column

```python
covid_19_df.rename(columns = {"area": "county_name"}, inplace=True,)
```

```python
covid_19_df.head()
```

#### Remove California, Unknown, and Out of State from county_name column

```python
covid_19_county_df = covid_19_df[covid_19_df["county_name"].str.contains("California|Unknown|Out of state")==False]
```

```python
covid_19_county_df.head()
```

#### Only include dataset from 2020 to 2021

```python
covid_19_county_df_2020_2021 = covid_19_county_df[covid_19_county_df["date"].str.contains("2021|2020")==True]
```

```python
covid_19_county_df_2020_2021.head()
```

#### Aggregate with group-by

```python
covid_group_county = covid_19_county_df_2020_2021.groupby(["county_name"]).agg({"deaths":"sum", "population": "mean"}).reset_index()
```

```python
covid_group_county.head()
```

#### Calculate mortality rate by culmination of deaths/population of county per 100k Population in 2020-201

```python
covid_group_county["covid_death_per_pop"]=(covid_group_county.deaths/covid_group_county.population)*100000
```

```python
covid_group_county.head()
```

#### Top 5 counties with highest mortality rates in CA

```python
covid_group_county.sort_values("covid_death_per_pop", ascending = False).head()
```

#### First 5 rows of GeoJSON file

```python
counties_gdf = gpd.read_file(
    "https://raw.githubusercontent.com/stiles/usc/main/data/geo/counties_simple.geojson"
)
```

```python
counties_gdf.head()
```

#### Plot of data

```python
counties_gdf.plot()
```

#### Merging gdf and value dataset

```python
merged_counties_covid_2022_gdf = pd.merge(counties_gdf, covid_group_county, on = "county_name")
```

```python
merged_counties_covid_2022_gdf.tail(15)
```

## Export geodataframe as GeoJSON

```python
merged_counties_covid_2022_gdf.to_file("data/counties_covid_death_2020_2021.geojson", driver="GeoJSON")
```
