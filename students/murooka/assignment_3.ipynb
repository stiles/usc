Import dataset

```python
covid_data=pd.read_csv("https://raw.githubusercontent.com/datadesk/california-coronavirus-data/master/cdph-county-cases-deaths.csv", dtype = {"fips":str})
```

1. Counting number of rows in the dataset

```python
len(covid_data)
```

```python
covid_election_date=covid_data[covid_data["date"]=="2021-09-14"].copy()
```

```python
covid_election_date.head()
```

```python
len(covid_election_date)
```

```python
covid_election_date["confirmed_rate"]=(covid_election_date["confirmed_cases"]/covid_election_date["population"])*1000
```

```python
covid_election_date.head()
```

2. Sorting my data


```python
covid_election_date.sort_values("confirmed_rate", ascending = False).head()
```
