#### Load python tools:

```python
import pandas as pd 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datadesk/california-coronavirus-data/master/cdph-vaccination-county-totals.csv')
```

### Explore the dataset

```python
df.head()
```

#### Numbers of rows in data

```python
len(df)
```

#### Sort by fully vaccinated individuals

```python
df.sort_values('county', ascending=False).head()
```

#### How many individuals are fully vaccinated?

```python
df.value_counts('fully_vaccinated')
```

#### Explore distributions

```python
df.describe(exclude=None)
```

#### Most recent

```python
latest_df = df[df["date"] == df["date"].max()]
```

```python
len(latest_df)
```

#### Columns

```python
essential_df = latest_df[["county", "fully_vaccinated_percent", "population"]]
```
