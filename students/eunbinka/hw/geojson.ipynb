## Geojson Homework
The Peregrine Falcon Observations Database is a product of the California Department of Fish and Wildlifes (CDFWs) Wildlife Branch and Biogeographic Data Branch. It was created by the Santa Cruz Predatory Bird Research Group. The records generally represent peregrine falcons (PEFA, Falco peregrinus) nesting, perching and flying. 

#### Importing Tools & Data

```python
%load_ext lab_black
```

```python
import pandas as pd
import geopandas as gpd
```

```python
pd.options.display.max_columns = 1000
pd.options.display.max_rows = 1000
```

```python
falcon = gpd.read_file("../data/geo/peregrine_falcon_observations.geojson")
```

#### Explore the Dataset

```python
falcon
```

```python
falcon.plot()
```

```python
include = [
    "Malibu Beach",
    "Long Beach",
    "Redondo Beach",
    "Newport Beach",
    "Laguna Beach",
    "Dana Point",
    "La Jolla",
]
```

```python
falcon_gdf = falcon[~falcon["QUADNAME"].isin(include)]
```

```python
falcon_gdf.plot()
```

```python
falcon_gdf.to_file("../data/geo/beach_cities_falcon_observations", driver="GeoJSON")
```
