```python
import pandas as pd
import altair as alt
```

```python
pd.options.display.max_columns = 1000
pd.options.display.max_rows = 1000
```

### Data

```python
demo = pd.read_csv(
    "https://raw.githubusercontent.com/stiles/usc/main/data/processed/vax_vote_demographics_ca_counties.csv"
)
```

```python
demo
```

#### Build a bar chart

```python
alt.Chart(demo).mark_bar().encode(x="name", y="mult_race")
```

#### Add more features: color, line width, title, size

```python
alt.Chart(demo).mark_bar(color="orange", size=12).encode(
    x="name", y="mult_race"
).properties(width=870, height=400, title="Multiracial population for each county")
```
