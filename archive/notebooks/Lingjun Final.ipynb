#### Import Vaccine Data

```python
src = pd.read_csv(
    "https://raw.githubusercontent.com/datadesk/california-coronavirus-data/master/cdph-vaccination-zipcode-totals.csv"
)
```

```python
df = src[src["fully_vaccinated_percent"] > 0.01].copy()
```

```python
df.dtypes
```

```python
df["fully_vaccinated_percent"] = df["fully_vaccinated_percent"].astype(int)
```

```python
LA_county = "Los Angeles"
LA_vacc = df[df.county == LA_county]
LA_vacc
```

#### Compare the latest data with back then April

```python
df["nudate"] = pd.to_datetime(df["date"])
```

```python
vacc_recent = LA_vacc[LA_vacc["date"] == LA_vacc["date"].max()]
```

```python
vacc_early = LA_vacc[LA_vacc["date"] == "2021-04-20"]
```

```python
vacc_recent[
    (vacc_recent.county == "Los Angeles")
    & (vacc_recent["fully_vaccinated_percent"] < 1)
].sort_values("fully_vaccinated_percent", ascending=False).head()
```

```python
vacc_early[
    (vacc_early.county == LA_county) & (vacc_early["fully_vaccinated_percent"] < 1)
].sort_values("fully_vaccinated_percent", ascending=False).head(5)
```

```python
vacc_early.groupby(["id"]).agg(
    {"fully_vaccinated_percent": "mean"}
).reset_index().sort_values("fully_vaccinated_percent", ascending=False).head(15)
```

```python
vacc_recent.groupby(["id"]).agg(
    {"fully_vaccinated_percent": "mean"}
).reset_index().sort_values("fully_vaccinated_percent", ascending=False).head(15)
```

#### Merge april vax with the demographic data

```python
April = pd.merge(data, vacc_early, left_on="zip", right_on="id", how="left",)
```

```python
April.drop(["id"], axis=1, inplace=True)
```

```python
len(April)
```

```python
April.head()
```

```python
LA_April = (
    April[April["county"] == "Los Angeles"]
    .sort_values("fully_vaccinated_percent", ascending=False)
    .head(5)
)
LA_April
```

#### Merge latest vax with the demographic data

```python
recent = pd.merge(data, vacc_recent, left_on="zip", right_on="id", how="left",)
```

```python
recent.drop(["id"], axis=1, inplace=True)
```

```python
len(recent)
```

```python
LA_recent = (
    LA_recent[LA_recent["county"] == "Los Angeles"]
    .sort_values("fully_vaccinated_percent", ascending=False)
    .head(5)
)
LA_recent
```

```python
LA_recent["white_percent"] = LA_recent["white"] / LA_recent["population"]
```
